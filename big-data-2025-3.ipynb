```python
#1. ENVIRONMENT SETUP
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings("ignore")
%matplotlib inline

# Set Seaborn style
sns.set(style="whitegrid")

#2. LOAD DATA
df = pd.read_csv(os.path.join(os.getcwd(), 'data', 'OnlineNewsPopularity.csv'))
df.columns = df.columns.str.strip()  # remove trailing whitespaces
df.drop('url', axis=1, inplace=True)

#3. BASIC DATA OVERVIEW
print("Shape:", df.shape)
display(df.head())
display(df.describe())

# Add log-transformed target & binary class
df['log_shares'] = np.log1p(df['shares'])
df['popular'] = (df['shares'] > df['shares'].median()).astype(int)

# Target variable and log-transformed target overview (side-by-side)
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.histplot(df['shares'], bins=100, kde=True)
plt.title('Distribution of Shares')
plt.xlabel('Shares')
plt.xscale('log')

plt.subplot(1, 2, 2)
sns.histplot(df['log_shares'], bins=100, kde=True, color='orange')
plt.title('Distribution of Log-Transformed Shares')
plt.xlabel('log_shares')

plt.tight_layout()
plt.show()
```

```python
# 4. MISSING VALUES CHECK
missing_counts = df.isnull().sum()
print("Missing values:\n", missing_counts[missing_counts > 0])
```

```python
# 5. UNIVARIATE ANALYSIS - NUMERICAL

# (1) Top 10 Features Most Correlated with 'shares'
top_corr = df.corr()['shares'].abs().sort_values(ascending=False)
top_features = [feat for feat in top_corr.index if feat not in ['shares', 'log_shares', 'popular']][:10]

print("Top 10 features most correlated with 'shares':")
print(top_features)

df[top_features].hist(figsize=(16, 10), bins=30)
plt.suptitle("Top 10 Most Correlated Features with 'shares'")
plt.tight_layout()
plt.show()

# (2) Low-Correlation Features (for contrast)
low_corr = top_corr[top_corr < 0.01].index.tolist()[:3]

print("\nExamples of weakly correlated features:")
print(low_corr)

df[low_corr].hist(figsize=(10, 5), bins=30)
plt.suptitle("Examples of Low-Correlation Features")
plt.tight_layout()
plt.show()
```

```python
# 6. UNIVARIATE ANALYSIS - CATEGORICAL (Channel + Day)
channel_cols = [col for col in df.columns if col.startswith('data_channel')]
day_cols = [col for col in df.columns if col.startswith('weekday_is')]

# Count plot for channel
df[channel_cols].sum().plot(kind='bar', title='Articles per Channel', figsize=(8,4))
plt.ylabel('Article Count')
plt.xticks(rotation=45)
plt.show()

# Count plot for days
df[day_cols].sum().plot(kind='bar', title='Articles per Day of Week', figsize=(8,4))
plt.ylabel('Article Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. BIVARIATE ANALYSIS - Correlation
plt.figure(figsize=(14, 12))
corr = df.corr()
sns.heatmap(corr[['shares']].sort_values(by='shares', ascending=False), annot=True, cmap='coolwarm')
plt.title("Correlation with 'shares'")
plt.show()
```

```python
# Box & Violin Plots for key features vs popularity
key_features = ['n_tokens_title', 'num_imgs', 'num_videos', 'average_token_length']

for feat in key_features:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.boxplot(x='popular', y=feat, data=df, ax=axes[0])
    sns.violinplot(x='popular', y=feat, data=df, ax=axes[1])
    axes[0].set_title(f'Boxplot of {feat}')
    axes[1].set_title(f'Violinplot of {feat}')
    plt.tight_layout()
    plt.show()
```

```python
# 8. MULTIVARIATE ANALYSIS - Shares by Channel and Day
df['weekday'] = df[day_cols].idxmax(axis=1)
df['channel'] = df[channel_cols].idxmax(axis=1)

plt.figure(figsize=(10, 5))
sns.boxplot(x='weekday', y='log_shares', data=df)
plt.xticks(rotation=45)
plt.title("Log Shares by Weekday")
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(x='channel', y='log_shares', data=df)
plt.xticks(rotation=45)
plt.title("Log Shares by Data Channel")
plt.show()
```

```python
# 9. FEATURE ENGINEERING - Engagement Score
df['engagement_score'] = (
    df['num_hrefs'] + df['num_self_hrefs'] + df['num_imgs'] +
    df['num_videos'] + df['num_keywords']
)

plt.figure(figsize=(8,4))
sns.kdeplot(data=df, x='engagement_score', hue='popular', fill=True)
plt.title("Engagement Score Distribution by Popularity")
plt.show()
```

```python
# 10. PREPROCESSING
from sklearn.preprocessing import StandardScaler

# target = 'popular'
# X = df.drop(['shares', 'log_shares', 'popular', 'weekday', 'channel'], axis=1)
# y = df['popular']
# OR
# target = 'shares' ('log_shares')
X = df.drop(['shares', 'log_shares', 'popular', 'weekday', 'channel'], axis=1)
y = df['log_shares']  

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Feature shape:", X_scaled.shape)
print("Target shape:", y.shape)
```

```python
# Convert X_scaled (numpy array) and y (Series) to DataFrame for saving
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)
y_df = pd.DataFrame(y)
# Optionally, join for a single DataFrame
preprocessed_df = X_scaled_df.join(y_df)
preprocessed_df.to_csv(os.path.join(os.getcwd(), 'data', 'preprocessed_data.csv'), index=False)
```
